## pspy tutorials



This directory holds several tutorials on how to use `pspy` API. 

List of tutorials:

-   [`tutorial_io`](tutorial_io.ipynb) shows how to read/write/plot SO map

-   [`tutorial_projection`](tutorial_projection.ipynb) explains how to project an `HEALPIX` map to a `CAR` pixellised map

-   [`tutorial_spectra_car_spin0and2`](tutorial_spectra_car_spin0and2.ipynb) shows how to generate spin0 and spin2 spectra for `CAR` pixellisation

-   [`tutorial_spectra_healpix_spin0and2`](tutorial_spectra_healpix_spin0and2.ipynb) shows how to generate spin0 and spin2 spectra for `HEALPIX` pixellisation
